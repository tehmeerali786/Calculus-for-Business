```python
def price_supply(p1, p2, s1, s2):
  m = round((p1-p2), 2)/round((s1-s2), 2)
  return round(m, 2)
```

```python
def price_demand(p1, p2, d1, d2):
  m = ((p1-p1)/(d1-d2))
  return round(m, 2)
```

```python
m_supply = price_supply(13.19, 12.59, 695, 595)
```

```python
m_demand = price_demand(12.59, 13.19, 650, 590)
```

```python
# Compound interest rate
def f(P, r, m, t):
  Indt5 = np.where(t==5)
  ex = (m*t)
  A = P*((1 + (r/m))**ex)
  return A, A[Indt5]
```

```python
t = np.arange(1, 21)
```

```python
len(t)
```

```python
A, A_5 = f(5000, .09, 365, t)
```

```python
len(A)
```

```python
print(f"Whole tweenty years money: {A}")
print(f"Firth years interest: {A_5}")
```

```python
np.column_stack((t, A))
```

```python
fig, ax = plt.subplots()
ax.plot(t, A, color='blue', label="f(x)")
ax.plot(5, A_5, marker='o', markersize=5, markerfacecolor='black')

ax.grid()
leg = ax.legend(loc='upper center')
plt.show()
```

```python
# Continuous compound interest
def f(P, r, t):
  t_5 = np.where(t==5)
  b = r*t
  A = P*(np.exp(b))
  return A, A[t_5]
```

```python
A, A_5 = f(5000, .09, t )
```

```python
print(f"Whole 20 years: {A}")
print(f"At the 5th year {A_5}")
```

```python
fig, ax = plt.subplots()
ax.plot(t, A, color='blue', label="f(x)")
ax.plot(5, A_5, marker='o', markersize=5, markerfacecolor='black')

ax.grid()
leg = ax.legend(loc='upper center')
plt.show()
```
